# Imports

```python
#Importing all of the packages that will be needed for this notebook
import pandas as pd
import numpy as np
import os
```

# Loading the Data

We are getting our dataset from the UCI Machine Learning data base. The download link is https://archive.ics.uci.edu/ml/machine-learning-databases/00373/drug_consumption.data

```python
#Our data file is saved in the raw_data directory of our repository
respondent_data = pd.read_csv('../raw_data/drug_consumption.data', header=None)
respondent_data
```

# Choosing the Columns that will be used

We can find a description of the dataset from the Drug Consumption page of the UCI Database (https://archive.ics.uci.edu/ml/datasets/Drug+consumption+%28quantified%29#)

From this page, we see that the dataset should contain 32 attributes and 1885 instances. Each instance represents an individual respondent, each attribute represents a different attribute, where one is the ID, 12 are personality attributes and the rest are drug use attributes. We can find descriptions of each attribute (now columns) from the above link. The loaded data does currently have 32 columns and 1885 rows, so we know this has loaded as expected.

```python
#Let's rename our variables
column_names = ['ID', 'age', 'gender', 'education', 'residing_country', 'race', 'n_score', 'e_score', 'o_score', 'a_score', 'c_score', 'i_score', 's_score', 'alcohol', 'amphet', 'amyl', 'benzos', 'caffeine', 'cannabis', 'chocolate', 'cocaine', 'crack', 'ecstasy', 'heroin', 'ketamine', 'legalh', 'lsd', 'meth', 'mushrooms', 'nicotine', 'semeron', 'vsa']
respondent_data.columns = column_names
respondent_data
```

Most of the columns are intuitive, except for the x_score columns. n_score represents neuroticism (which is one's disposition to experience negative emotions such as anger or anxiety), e_score represents extraversion (which represents how outgoing or social an individual is), o_score represents one's openness to experience or trying new things, a_score represents agreeableness or one's ability to put others' needs over their own (can be associated with altruistic or empathetic behavior), c_score representing conscientiousness or the tendency to adhere to norms/rules and do what is right. These are widely considered to be the "Big Five" personality traits in the world of psychology. This data also includes i_score, which represents impulsivity or tendency to act without thinking/on impulse. This is quite related to conscientiousness, as the more impulsive one is, the less conscientious they tend to be. Finally, an s-score is included which represents sensation-seeking or one's tendency to search for experiences and feelings.

Since the ID column does not give us any useful information on the respondents, we will get rid of this column. 

```python
#dropping unnecessary columns
respondent_reduced = respondent_data.drop(columns = ['ID'])
respondent_reduced
```

Note that we did not get rid of the semeron column. This column represents consumption of a fictitious drug called semeron, and can be useful to identify respondents who may be untruthful or those who are overclaiming their drug consumption. Realisitically, this should be CL0 (which represents never used) for every respondent, and for those who do claim otherwise, we may need to get rid of their data.

# Initial Exploration

We know that the semeron column was used as a test to see how truthful the respondents were being. Since semeron is a fictitious drug, we should get rid of any respondent/row that claims to have taken semeron in the past, i.e. any row where semeron column value is not CL0.

```python
#Find how many row have value not equal to CL0 in semeron column
df_drugs[df_drugs['semeron'] != 'CL0']
```

```python
#Only 8 out of 1885 respondents claimed to have taken semeron, so we should remove these rows 
df_drugs = df_drugs[df_drugs['semeron'] == 'CL0']
df_drugs
```

```python
#lets now drop the semeron column since it is no longer needed
df_drugs = df_drugs.drop(columns = 'semeron')
df_drugs
```

```python
#Lets also see if we have any N/A values
df_drugs.isna().sum()
```
